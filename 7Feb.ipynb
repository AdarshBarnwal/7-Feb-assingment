```python
Q1.
Ans:-
# create a functin
def check_password(password):
    
    password=input("Enter your password")    #password input
    
    up_case=0
    lw_case=0
    special_char=0
    num=0
    
    for i in password:
        if i.islower():
            lw_case=lw_case+1
        elif i.isupper():
            up_case=up_case+1
        elif i == "@" or i =='!' i =='#' i=='&' i=='*' i=='_':
            special_char=special_char+1
        elif i.isnumeric():
            num=num+1
#condition check
    if(lw_case>=2 and up_case>=2 and num>=1 and special_char>=3 and len(password)>=10):  
        return"valid password"
    else:
        return"invalid password"
#function call
check_password(password)
```
